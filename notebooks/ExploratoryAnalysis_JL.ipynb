## Preprocessing Data

1. Consolidate all crimes under 500 counts into "Other Offense" Category

```python
df[' PRIMARY DESCRIPTION'].value_counts()
```

```python
df['Offenses'] = df[' PRIMARY DESCRIPTION']
```

```python
value_counts = df['Offenses'].value_counts()
```

```python
to_replace = value_counts[value_counts<500].index
```

```python
df['Offenses'] = df['Offenses'].replace(to_replace, "OTHER OFFENSE")
```

```python
df['Offenses'].value_counts()
```

2. Merge similar offenses:
- Criminal Sexual Assault -> Sex Offense
- Public Peace Violation --> Interference with Public Officer --> new category: Public Order
  

```python
df['Offenses'] = df['Offenses'].replace({
    "PUBLIC PEACE VIOLATION":"PUBLIC ORDER",
    "INTERFERENCE WITH PUBLIC OFFICER":"PUBLIC ORDER",
    "CRIMINAL SEXUAL ASSAULT":"SEX OFFENSE"
    })
```
